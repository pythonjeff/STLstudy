```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mticker
import numpy as np
```

```python
city_pop = pd.read_csv("Desktop/MOSSPOP.csv")
msa_pop = pd.read_csv("Desktop/STLPOP.csv")
```

```python
msa_pop.head(5)
```

```python
city_pop.replace([float('inf'), -float('inf')], pd.NA, inplace=True)
```

```python
city_pop['MOSSPOP'] = city_pop['MOSSPOP'] * 1000
msa_pop['STLPOP'] = msa_pop['STLPOP'] * 1000
```

```python
city_pop['DATE'] = pd.to_datetime(city_pop['DATE'])
city_pop['YEAR'] = city_pop['DATE'].dt.year
city_pop.set_index('YEAR', inplace=True)
city_pop.drop(columns=['DATE'], inplace=True)
```

```python
msa_pop['DATE'] = pd.to_datetime(msa_pop['DATE'])
msa_pop['YEAR'] = msa_pop['DATE'].dt.year
msa_pop.set_index('YEAR', inplace=True)
msa_pop.drop(columns=['DATE'], inplace=True)
```

```python
msa_pop.head(5)
```

```python
city_pop.head(5)
```

```python
sns.lineplot(data=city_pop, x='YEAR', y='MOSSPOP')
plt.ylabel("population in hundred thousands")
plt.title("St. Louis CITY")
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
plt.show()
```

```python
sns.lineplot(data=msa_pop, x='YEAR', y='STLPOP')
plt.ylabel("MSA Population")
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
plt.title("St. Louis MSA")
plt.show()
```
